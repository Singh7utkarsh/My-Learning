```python
import pandas as pd
pd.set_option("display.max_columns",400)
```

```python
gg = pd.read_csv(r"C:\Users\HP\Downloads\Sample for Reportt.csv")
gg = gg[~gg['Case Reason'].isna()]
gg
```

```python
##Open Cases
gg1 = gg[(gg['Status']=='Open') & (gg['Mailing State/Province']=='JHARKHAND')]
print(len(gg1))
gg1
```

```python
## Open Cases Summary
gg2 = pd.DataFrame(gg1.groupby('Assigned Service Provider')['Case Reason'].count()).reset_index()
gg2.loc[len(gg2)] = gg2.sum(numeric_only = True)
gg2 = gg2.fillna('Total')
gg2.columns = ['Assigned Service Provider', 'Open Cases']
gg2
```

```python
## NPS
z = ['Partially Closed','Closed']
b = [9,10]
c = [7,8]
d = [1,2,3,4,5,6]

gg3 = gg[gg['Status'].isin(z)]
gg3.loc[gg3['NPS overall rating'].isin(b), "NPS Score"] = 1
gg3.loc[gg3['NPS overall rating'].isin(c), "NPS Score"] = 0
gg3.loc[gg3['NPS overall rating'].isin(d), "NPS Score"] = -1
gg3 = gg3.filter(['Assigned Service Provider','NPS Score'])
gg3 = gg3[~gg3['NPS Score'].isna()]
gg3 = gg3.groupby('Assigned Service Provider')['NPS Score'].agg(['sum','count']).reset_index()
gg3.loc[len(gg3)] = gg3.sum(numeric_only = True)
gg3 = gg3.fillna('Total')
gg3['NPS'] = (100*gg3['sum'])/gg3['count']
gg3['NPS'] = gg3['NPS'].round()
gg3
```

```python
## Response Rate
gg4 = gg[gg['Status'].isin(z)]
gg4.loc[gg4['NPS overall rating'].isna(),"Response"] = 0
gg4.loc[~gg4['NPS overall rating'].isna(),"Response"] = 1
gg4 = gg4.filter(['Assigned Service Provider','Response'])
gg4 = gg4.groupby('Assigned Service Provider')['Response'].agg(['sum','count']).reset_index()
gg4.loc[len(gg4)] = gg4.sum(numeric_only = True)
gg4 = gg4.fillna('Total')
gg4['Response Rate'] = (100*gg4['sum'])/gg4['count']
gg4['Response Rate'] = gg4['Response Rate'].round()
gg4
```

```python
## Same Day Closure
gg5 = gg.filter(['Assigned Service Provider','24 hrs'])
gg5 = gg5.groupby('Assigned Service Provider')['24 hrs'].agg(['sum','count']).reset_index()
gg5.loc[len(gg5)] = gg5.sum(numeric_only = True)
gg5 = gg5.fillna('Total')
gg5['SDC'] = (100*gg5['sum'])/gg5['count']
gg5['SDC'] = gg5['SDC'].round()
gg5
```

```python
## Same Day Closure
gg6 = gg[gg['Status']=='Closed']
gg6 = gg6.filter(['Assigned Service Provider','72hrs Full Closure'])
gg6 = gg6.groupby('Assigned Service Provider')['72hrs Full Closure'].agg(['sum','count']).reset_index()
gg6.loc[len(gg6)] = gg6.sum(numeric_only = True)
gg6 = gg6.fillna('Total')
gg6['72 hrs'] = (100*gg6['sum'])/gg6['count']
gg6['72 hrs'] = gg6['72 hrs'].round()
gg6
```

```python
with pd.ExcelWriter('KPI.xlsx') as writer:
    gg.to_excel(writer, sheet_name = 'Raw data', index = False)
    gg1.to_excel(writer, sheet_name = 'Open Cases', index = False)
    gg2.to_excel(writer, sheet_name = 'Open Cases Summary', index = False)
    gg3.to_excel(writer, sheet_name = 'NPS', index = False)
    gg4.to_excel(writer, sheet_name = 'Response Rate', index = False)
    gg5.to_excel(writer, sheet_name = 'SDC', index = False)
    gg6.to_excel(writer, sheet_name = '72 hrs closure', index = False)
```
